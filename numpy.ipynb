Array Operations 

```python
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

c = a - b 
d = a * b 

print(c)
print(d)
```

```python
# conversion of a list of farenheit to celcius 

farenheit = np.array([0, -10, -5, -15, 0])
celcius = (farenheit - 31) * (5/9)

print(celcius)
```

```python
celcius > -20
```

```python
farenheit >= 0
```

```python
A = np.array([[1, 1],[0, 1]])
B = np.array([[2, 0], [3, 4]])
# elementwise product
print(A * B)

# matrix product 
print(A @ B)
```

```python
array1 = np.array([[1, 2, 3], [4, 5, 6]])
print(array1.dtype)
```

```python
array2 = np.array([[10.5, 5.1, 64.4], [61.7, 21.0, 12.0]])
print(array2.dtype)
```

```python
array3 = array1 + array2 
print(array3) 
print(array3.dtype)
```

```python
print(f"this is the min {array3.min()}")
print(f"this is the max {array3.max()}")
print(f"this is the mean {array3.mean()}")
print(f"this is the sum {array3.sum()}")
```

```python
from PIL import Image 
from IPython.display import display

im = Image.open('ahmethan.jpg')
display(im)
```

```python
mask = np.full(array.shape, 255)
mask
```

```python
modified_array = array - mask 
modified_array = modified_array*-1
modified_array = modified_array.astype(np.uint8)
modified_array
```
