```python
import os
import numpy as np
from PIL import Image


def load_custom_dataset(image_paths, annotation_file, bbox ,**kwargs):
     # Load as dictionary
    
    samples = []
    for image_path in range(len(image_paths)):
        image = Image.open(image_paths[image_path])
        image_name, _ = os.path.splitext(os.path.basename(image_paths[image_path]))  # Extract image name without extension
        
        annotation_key = image_name + '.txt'
        
        if annotation_key in annotation_file:
            annotation = bbox[image_path]
            if isinstance(annotation, np.ndarray) and annotation.size == 1:
                annotation = annotation.item()  # Convert to Python scalar if necessary
            samples.append((image, annotation))
        else:
            print(f"Warning: Annotation not found for image {image_name}")
    
    return samples



# for i, image_path in enumerate(images_path):
#     new_dir = '/kaggle/input/dataset-vehicle1/images'+'/'+image_path
#     image = cv2.imread(new_dir)
#     label = labels[i] 
#     elif i >= 5694:
#         image = cv2.imread(image_path)
#         label = labels[i+1]  # Get the corresponding label from the loaded labels array

#     labeled_images[image_path] = {'image': image, 'label': label}
```
